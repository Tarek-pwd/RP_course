# ROS Package Dependencies

ROS packages have two types of dependencies:

1. **Build Dependencies**
2. **Run Dependencies**

The `rosdep` tool checks for a package's missing dependencies, downloads them, and installs them.

---

## Checking for Missing Dependencies

To check for missing dependencies in a ROS package:

```bash
$ rosdep check <package name>
```

**Note**: For this command to work, the workspace must first be sourced:

```bash
$ source devel/setup.bash
```

This command provides a list of the system dependencies that are missing and where to get them.

---

## Installing Dependencies with rosdep

To have `rosdep` install packages, invoke the following command from the root of the catkin workspace:

```bash
$ rosdep install -i <package name>
```
