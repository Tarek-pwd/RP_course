## Placing the Robot in a Custom Environment

![img](lesson4p11_img1.png)

### Adding the World File

So far, you created a robot model from scratch, added sensors to it to visualize its surroundings, and developed a package to launch the robot in a simulated environment. That's a real accomplishment!

But you haven’t yet placed the robot in an environment. Let’s house it inside the world you built in the Build My World project.

#### Steps:
1. **Copy the `<yourname>.world` File**
   - Copy the `<yourname>.world` file from the world directory of the Build My World project and paste it in the `worlds` directory of `my_robot`.
   - Inside your package’s `worlds` directory, you should now see two files:
     - `empty.world` (created earlier).
     - `<yourname>.world` (just added).
   - Feel free to delete the `empty.world` file. We won’t need it anymore.

---

### Launch the World

1. **Edit the `world.launch` File**
   - Add a reference to the `<yourname>.world` file that you just added. Open the `world.launch` file and edit this line:

     ```xml
     <arg name="world_file" default="$(find my_robot)/worlds/empty.world"/>
     ```

   - Replace it with:

     ```xml
     <arg name="world_file" default="$(find my_robot)/worlds/<yourname>.world"/>
     ```

2. **Launch the World**
   - Run the following commands to launch the world:

     ```bash
     $ cd ~/catkin_ws/
     $ source devel/setup.bash
     $ roslaunch my_robot world.launch
     ```

### Initialize the Robot’s Position and Orientation

As you can see, the robot’s initial position might be outside of the world! To fix this, you need to change the robot’s initial pose: its position and orientation.

1. **Edit the `world.launch` File**
   - Locate and modify the following lines in `world.launch`:

     ```xml
     <!-- Robot pose -->
     <arg name="x" default="0"/>
     <arg name="y" default="0"/>
     <arg name="z" default="0"/>
     <arg name="roll" default="0"/>
     <arg name="pitch" default="0"/>
     <arg name="yaw" default="0"/>
     ```

2. **Adjust Pose Values**
   - The best way to figure out these numbers is to:
     - Change the robot’s position and orientation within Gazebo.
     - Record its pose.
     - Update the `world.launch` file with the new values.
