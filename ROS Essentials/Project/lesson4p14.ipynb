## Adding a White Ball to the Gazebo World

Before you proceed to code the `process_image` client node, you have to model a white ball and place it in your Gazebo World scene.

After modeling the white ball, you'll control its position in Gazebo by placing it at different positions in front of the robot’s camera. The `process_image` client node will be responsible for analyzing the robot’s image and requesting services from the server `drive_bot` node to drive the robot towards it.

---

### Model Editor

Here’s a reminder of how to open the model editor:

```bash
$ gazebo # then Edit-> Model Editor
```

---

### Insert and Configure the Sphere

1. **Insert Sphere:**
   - Under the simple shapes menu of the Model Editor tool, click on a sphere and insert it anywhere in the scene.

2. **Edit Size:**
   - Double-click on the sphere and change its radius to `0.1` in both Visual and Collision.

3. **Change Color:**
   - To change the ball’s color to white, set its Visual Ambient, Diffuse, Specular, and Emissive RGBA values to `1`.

![img](lesson4p14_img1.png)

---

### Save and Insert the Ball

1. **Save:**
   - Save the white ball model as `my_ball` under the `/home` directory.
   - Exit the Model Editor tool and return to the Gazebo main world.

2. **Insert Ball:**
   - Back in the Gazebo main world, click on “Insert” and drop the white ball anywhere in the scene.

![img](lesson4p14_img2.png)

---

### Inserting the Ball into the World

1. **Open the World File:**
   - Open your world file using the following command:

     ```bash
     $ gazebo <path_to_your_world_file>
     ```

   - This opens the world without spawning your robot.

2. **Insert the Ball:**
   - Insert the `my_ball` model anywhere inside your world. The `my_ball` model can be found under the Insert tab.

![img](lesson4p14_img3.png)

3. **Save:**
   - Place the white ball anywhere outside of your building structure so that the robot cannot see it initially.
   - Save a copy of this new world under `~/catkin_ws/src/my_robot/worlds` by replacing your old `<yourname>.world` file.

---

### Verify the World

Whenever you launch this newly saved world, you should be able to see your building environment, and in addition, the white ball.
