# VM Workspace Access and Overview

## VM Workspace - Access

Follow the instructions below to access your VM workspace:

1. Open the left menu on the current page and scroll to the bottom of the menu.
2. Click the menu item **Launch Virtual Machine** at the bottom.
3. A pop-up will appear with a button **Open Cloud Console**. Click on the **Open Cloud Console** button.

![img](lesson1p9_img1.jpeg)

- When you click on the **Open Cloud Console** button, the VM workspace will open in a new browser tab. Ensure that you have not enabled the pop-up blocker for [https://learn.udacity.com/](https://learn.udacity.com/).
- The VM will start within 3-4 minutes. This is the average boot-up time for the machine.
- Review the information at the top bar when the VM loads. The top bar will display the VM status, duration used out of the total lifetime duration, current session timer, and start/extend/end/reset buttons. The screenshot below highlights the icons in the top bar of a VM workspace.

## VM Workspace - Overview

The VM workspace is an **Ubuntu 20.04 machine with NVIDIA Tesla T4 GPU support**. It includes the necessary tools for exercises and projects, such as:

- **CARLA simulator 0.9.9.4**
- **NICE DCV Server** (includes Nvidia drivers and CUDA libraries for Tesla T4 GPU)
- **C++**
- **Git**
- **OpenCV**
- **CMake** and **Make**
- **VSCode**
- **Eigen Library for C++**
- **Point Cloud Library**
- **Python3 and Pip**
- **ROS**

You can install additional tools if needed for exercises or projects.

### Restrictions and Best Practices

- The VM workspace is a **sandbox environment** with a limited session duration.
- After the session expires, the VM terminates.
- You must reload the classroom page to start a new session, but previous work will not be preserved.

#### Best Practices

1. **Save your edits** to a GitHub repository to ensure you can pull them in a future session.
2. **Session timer and lifetime duration**:
   - Each session lasts up to 4 hours.
   - Total lifetime duration is limited (e.g., 40 hours).
   - Extensions allow an extra hour up to four times per session (e.g., 8 hours total).
3. Save your work intermittently to avoid data loss.
4. Click **End Lab** when not using the VM to preserve runtime.

#### Important Notes

- Misusing the VM can result in revoked access or deactivation of your Udacity account.
- For assistance, file a support ticket via the link at the bottom-right of the classroom page.
