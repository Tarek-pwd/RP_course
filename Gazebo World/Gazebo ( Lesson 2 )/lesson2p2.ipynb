To help you get started, I’ve included a sample simulation. Reviewing the sample's aspects and directory structure will help you build your own world in Gazebo.

> **Note**: The sample simulation is provided only for reference, and you have visual access to it. It is not included in the Udacity Workspace or any other location.

## Project Aspects

Inside the Gazebo world, the following elements are included:

1. **Udacity Office**:
   - A building model created using the **Building Editor** tool in Gazebo.
   - Includes detailed features and colors.

2. **Green Humanoid Robot**:
   - Two instances of a model designed using the **Model Editor** tool in Gazebo.

3. **Tables**:
   - A model imported from the Gazebo online library.

4. **Terminal**:
   - A welcome message printed to the terminal using a world plugin.

```
.Project1                          # Build My World Project 
├── model                          # Model files 
│   ├── Building
│   │   ├── model.config
│   │   ├── model.sdf
│   ├── HumanoidRobot
│   │   ├── model.config
│   │   ├── model.sdf
├── script                         # Gazebo World plugin C++ script      
│   ├── welcome_message.cpp
├── world                          # Gazebo main World containing models 
│   ├── UdacityOffice.world
├── CMakeLists.txt                 # Link libraries 
└──
```

## Your Task

### Requirements

Your project should:
1. Include the same aspects as the sample project.
2. Follow the directory structure outlined above.

### Steps:

1. **Choose a Location**:
   - Select a place to deploy your robot, such as:
     - Your apartment
     - Your office
     - Your favorite restaurant

2. **Create a Custom Model**:
   - Design any model of your choice using the **Model Editor** tool in Gazebo.

3. **Import a Model**:
   - Select and import a model from the Gazebo online library.

4. **Design a Building**:
   - Use the **Building Editor** tool to create a structure representing your chosen location.

5. **Add a Welcome Plugin**:
   - Write a world plugin (e.g., `welcome_message.cpp`) to print a custom message in the terminal when the simulation runs.

By following these instructions, you can build a unique simulation environment in Gazebo tailored to your chosen location.
