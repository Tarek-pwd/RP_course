# Gazebo GUI Overview

The Gazebo GUI is divided into four major sections:

1. **Scene**
2. **Side Panel**
3. **Toolbars**
4. **Menu**

## Scene

The **Scene** is where you will spend most of your time, whether creating a simulation or running one. While you can use a trackpad to navigate inside the scene, a mouse is highly recommended.

- **Panning**: Press the left mouse button and drag.
- **Rotating**: Hold down `SHIFT` while dragging the left mouse button.
- **Zooming**: Use the mouse scroll or press and drag the right mouse button (RMB).

## Side Panel

The **Side Panel** on the left consists of three tabs:

### World

- Displays the lights and models currently in the scene.
- Allows viewing and editing of basic model parameters like position and orientation.
- Provides access to physics options, such as gravity and magnetic field, via the **Physics** option.
- The **GUI** option gives access to the default camera view angle and pose.

### Insert

- Contains objects (models) to add to the simulation scene.
- Left-click to expand or collapse a list of models.
- To place an object:
  1. Left-click the object under the Insert tab.
  2. Bind the object to your mouse cursor.
  3. Place it in the scene by left-clicking at the desired location.

### Layers

- An optional feature; this tab is usually empty.

## Top Toolbar

The **Top Toolbar** provides quick access to cursor types, geometric shapes, and views.

### Cursor Modes

1. **Select Mode**: The default cursor mode for navigating the scene.
2. **Translate Mode**: Adjust an object's position directly in the scene using axis-based dragging.
3. **Rotate Mode**: Change the orientation of an object.
4. **Scale Mode**: Modify the scale (size) of any model.

### Additional Tools

- **Undo/Redo**: Revert or redo actions.
- **Simple Shapes**: Add basic 3D models like cubes, spheres, or cylinders.
- **Lights**: Insert light sources, such as spotlight, point light, or directional light.
- **Copy/Paste**: Duplicate models using these tools or `Ctrl+C` and `Ctrl+V`.
- **Align**: Align models along one of the three principal axes.
- **Change View**: Switch between perspectives like top, side, front, or bottom views.

## Bottom Toolbar

The **Bottom Toolbar** includes:

- **Play/Pause Button**: Pause the simulation and move objects around.
- **Simulation Data**: Displays simulation time, real time, and their ratio.
- **Frames-per-Second Counter**: Monitors system performance for the scene.

## Menu

The **Menu** duplicates some toolbar options and right-click context menu options in the scene. Key features include:

- **Edit**: Switch to the Building Editor to design buildings or the Model Editor to create models.
